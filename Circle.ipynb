```python
from billiards import *
```

```python
def circle(s):
    s *= 2*np.pi
    return np.array([np.cos(s),np.sin(s)])
```

```python
x = np.array([.1, 0])
v = np.array([.1,-.2])
N = 500

ball = Billiards(circle, x, v, poincare = True)
for i in range(N):
    ball.bounce()
```

```python
ball.plot_path()
```

```python
ball.plot_poincare()
```
