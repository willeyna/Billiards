```python
from billiards import *
```

```python
x = np.array([.1,.1])
v = np.array([1,0])

N = 200
X = [x]
V = [v]
#bounce N times and plot trajectories
for i in range(N):
    new = bounce(X[i], V[i], circle, d_circle)
    X.append(new[0])
    V.append(new[1])
X = np.array(X)
V = np.array(V)
```

```python
fig = plt.figure(figsize = (15,5))

#plot circlular border
theta = np.linspace(0, 2*np.pi, 1000)
plt.plot(np.cos(theta), np.sin(theta), color = 'black')
plt.axis('equal')

for i in range(len(X)-1):
    plt.plot([X[i][0], X[i+1][0]], [X[i][1], X[i+1][1]], color = 'red', alpha = (i+1)/len(X))
```

```python
fig = plt.figure(figsize = (15,5))
plt.plot(X[:,1])
plt.xlabel("Bounce Number")
plt.ylabel("y-Value of ball hit")
```

Super periodic and predictable motion, obvious from symmetric pattern made by the ball's path
