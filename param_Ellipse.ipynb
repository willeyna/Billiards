```python
x = np.array([1/np.sqrt(2), 0])
v = np.array([0,1])

N = 10000
X = [x]
V = [v]
P = []
#bounce N times and plot trajectories
for i in range(N):
    new = param_bounce(X[i], V[i], ellipse, tol = 1e-4, poincare = True)
    X.append(new[0])
    V.append(new[1])
    P.append(np.array([new[2], new[3]])) 
X = np.array(X)
V = np.array(V)
P = np.array(P)
```

```python
fig = plt.figure(figsize = (15,5))

#plot elliptical border
a = np.linspace(-1, 1, 1000)
y_minus = -np.sqrt(1/2 - (a**2)/2)
y_plus = np.sqrt(1/2 - (a**2)/2)
plt.plot(np.concatenate([a,a[::-1]]), np.concatenate([y_minus,y_plus[::-1]]), color = 'black')
plt.axis('equal')

for i in range(len(X)-1):
    plt.plot([X[i][0], X[i+1][0]], [X[i][1], X[i+1][1]], color = 'red', alpha = (i+1)/len(X))
```

```python
fig = plt.figure(figsize = (10,10))

delt = .1
cmap = np.arange(P.shape[0])
plt.scatter(P[:,0], P[:,1], cmap = 'coolwarm', c = cmap, alpha = .5, s = 3)
plt.xlim(0-delt,1+delt)
plt.ylim(0-delt,1+delt)
plt.xlabel('Location on Curve')
plt.ylabel("Tangential Velocity")
plt.title(f"Billiards Poincare Section x=({x[0]}, {x[1]}),  v=({v[0]}, {v[1]})", fontsize = 14)
```
