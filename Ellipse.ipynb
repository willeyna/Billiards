```python
from billiards import *
```

```python
x = np.array([1/np.sqrt(2),0])
v = np.array([-.1,-.9])

N = 200
X = [x]
V = [v]
#bounce N times and plot trajectories
for i in range(N):
    new = bounce(X[i], V[i], ellipse, d_ellipse)
    X.append(new[0])
    V.append(new[1])
X = np.array(X)
V = np.array(V)
```

```python
fig = plt.figure(figsize = (15,5))

#plot elliptical border
a = np.linspace(-1, 1, 1000)
y_minus = -np.sqrt(1/2 - (a**2)/2)
y_plus = np.sqrt(1/2 - (a**2)/2)
plt.plot(np.concatenate([a,a[::-1]]), np.concatenate([y_minus,y_plus[::-1]]), color = 'black')
plt.axis('equal')

for i in range(len(X)-1):
    plt.plot([X[i][0], X[i+1][0]], [X[i][1], X[i+1][1]], color = 'red', alpha = (i+1)/len(X))
```

```python
fig = plt.figure(figsize = (15,5))
plt.plot(X[:,1])
plt.xlabel("Bounce Number")
plt.ylabel("y-Value of ball hit")
```

Motion doesn't seem entriely chaotic, definitely a periodicity.. next up is Poincare section I suppose!
